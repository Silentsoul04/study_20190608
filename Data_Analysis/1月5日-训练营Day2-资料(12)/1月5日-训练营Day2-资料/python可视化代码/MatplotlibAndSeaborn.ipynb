```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import os
from matplotlib.font_manager import FontProperties
```

```python
# 散点图
def scatter():
	# 数据准备
	N = 500
	x = np.random.randn(N)
	y = np.random.randn(N)
	# 用Matplotlib画散点图
	plt.scatter(x, y,marker='x')
	plt.show()
	# 用Seaborn画散点图
	df = pd.DataFrame({'x': x, 'y': y})
	sns.jointplot(x="x", y="y", data=df, kind='scatter');
	plt.show()
```

```python
scatter()
```

```python
# 折线图
def line_chart():
	# 数据准备
	x = [1900, 1901, 1902, 1903, 1904, 1905, 1906, 1907, 1908, 1909, 1910]
	y = [265, 323, 136, 220, 305, 350, 419, 450, 560, 720, 830]
	# 使用Matplotlib画折线图
	plt.plot(x, y)
	plt.show()
	# 使用Seaborn画折线图
	df = pd.DataFrame({'x': x, 'y': y})
	sns.lineplot(x="x", y="y", data=df)
	plt.show()
```

```python
line_chart()
```

```python
# 条形图
def bar_chart():
	# 数据准备
	x = ['c1', 'c2', 'c3', 'c4']
	y = [15, 18, 5, 26]
	# 用Matplotlib画条形图
	plt.bar(x, y)
	plt.show()
	# 用Seaborn画条形图
	sns.barplot(x, y)
	plt.show()
```

```python
bar_chart()
```

```python
# 箱线图
def box_plots():
	# 数据准备
	# 生成0-1之间的20*4维度数据
	data=np.random.normal(size=(10,4))
	lables = ['A','B','C','D']
	# 用Matplotlib画箱线图
	plt.boxplot(data,labels=lables)
	plt.show()
	# 用Seaborn画箱线图
	df = pd.DataFrame(data, columns=lables)
	sns.boxplot(data=df)
	plt.show()
```

```python
box_plots()
```

```python
# 蜘蛛图
def spider_chart():
	# 数据准备
	labels=np.array([u"推进","KDA",u"生存",u"团战",u"发育",u"输出"])
	stats=[76, 58, 67, 97, 86, 58]
	# 画图数据准备，角度、状态值
	angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
	stats=np.concatenate((stats,[stats[0]]))
	angles=np.concatenate((angles,[angles[0]]))
	# 用Matplotlib画蜘蛛图
	fig = plt.figure()
	ax = fig.add_subplot(111, polar=True)
	ax.plot(angles, stats, 'o-', linewidth=2)
	ax.fill(angles, stats, alpha=0.25)
	# 设置中文字体
	font = FontProperties(fname=r"simheittf.ttf", size=14)
	ax.set_thetagrids(angles * 180/np.pi, labels, FontProperties=font)
	plt.show()
```

```python
spider_chart()
```

```python
# 二元变量分布图
def jointplot():
	# 数据准备
	# flights = sns.load_dataset("flights")
	flights = pd.read_csv("flights.csv")
	# 用Seaborn画二元变量分布图（散点图，核密度图，Hexbin图）
	sns.jointplot(x="year", y="passengers", data=flights, kind='scatter')
	sns.jointplot(x="year", y="passengers", data=flights, kind='kde')
	sns.jointplot(x="year", y="passengers", data=flights, kind='hex')
	plt.show()
```

```python
jointplot()
```

```python
# 成对关系图
def pairplot():
	# 数据准备
	# flights = sns.load_dataset('flights')
	flights = pd.read_csv("flights.csv")
	# 用Seaborn画成对关系

	sns.pairplot(flights)
	plt.show()
```

```python
pairplot()
```
